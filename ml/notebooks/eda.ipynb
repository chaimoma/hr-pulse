```python
#examine Salary Estimate column
print(df['Salary Estimate'].value_counts().head(10))
```

```python
#clean text columns
df['Job Title'] = df['Job Title'].str.strip().str.lower()
df['Job Description'] = df['Job Description'].str.strip().str.lower()
df['Job Description'] = df['Job Description'].str.replace('\n', ' ', regex=True)
df['Job Description'] = df['Job Description'].str.replace('\s+', ' ', regex=True)
#standardize text for NLP or DB ingestion.
```

```python
#extract numeric salary
#remove $ and K, extract min and max
salary_range = df['Salary Estimate'].str.extract(r'\$?(\d+)K?-\$?(\d+)K?')
df['salary_min'] = salary_range[0].astype(float)
df['salary_max'] = salary_range[1].astype(float)
df['salary_avg'] = (df['salary_min'] + df['salary_max']) / 2
```

```python
# keep useful columns for ML
df_cleaned = df[
    [
        'Job Title',
        'Job Description',
        'Rating',
        'Location',
        'Size',
        'Industry',
        'Sector',
        'Founded',
        'salary_avg'
    ]
]
# Rename columns
df_cleaned.columns = [
    'job_title',
    'job_description',
    'rating',
    'location',
    'size',
    'industry',
    'sector',
    'founded',
    'salary_avg'
]
```

```python
#save
df_cleaned.to_csv("../../data/data_cleaned.csv", index=False)
```
